## NUMPY ARRAY BASICS##

```python
import numpy as np
import time
```

 ## CREATING ARRAYS FROM NUMPY#$
 

```python
arr_1d = np.array([1,2,3,4,5,6])
arr_2d = np.array([[1,2,3],[4,5,6]])
print("1D array: ",arr_1d)
print("2D array: ",arr_2d)
```

## list V/S  numpy arrays

```python
list1 = [1,2,3]
print("list multiplication:", list1 *2)

arry = np.array([1,2,3])
print("array multiplication:", arry *2)
```

##CECKING THE EFFICIENCY OF THE LISTS AND ARRAYS##

```python
start = time.time()
l1= [i*2 for i in range(10000)]
print("\n operation time    :", time.time()-start)


start = time.time()
a1=np.arange(10000)*2
print("\n operation time    :", time.time()-start)
```

##TYPES OF OTHER ARRAYS##

```python
#Zeros Matrix
zeros=np.zeros((3, 4))
print("\nArray of zeros: \n",zeros)

ones =np.ones((4,4))
print("\nArray of ones: \n",ones)

full= np.full((4,4), 7) 
print("\nArray of fulls: \n",full)

random = np.random.random((3,3))
print("\nArray of randoms: \n",random)

sequence = np.arange(0,100,5)
print("\nArray of sequences: \n",sequence)
```

## Vector, Matrix, Tensor ##

```python
vector= np.array([1,2,3])
print("vector: ", vector)

matrix = np.array([[1,2,3],[1,2,3]])
print("matrix:\n ", matrix)

tensor = np.array([[[1,2],[2,3]],
                    [[4,5],[6,7]],
                   [[8,9],[9,10]]])
print("Tensor:\n ", tensor)
```

## properties of the array##

```python
x1= np.array([[1,2,3],[4,5,6]] )
print("Shape of an array: ", x1.shape)
print("Dimensions: ",x1.ndim)
print("size: ",x1.size)
print("DTYPE: ",x1.dtype)
```

## Array Reshaping

```python
arr = np.arange(12)
print("Array: ",arr)

reshaping = arr.reshape((3,4))
print("Reshaped: \n", reshaping)

flattened = reshaping.flatten()
print("Flatenned Array: \n",flattened)

#ravel shows you view, instead of copy
ravled = reshaping.ravel()
print("Raveled: \n", ravled)

transpose = reshaping.transpose()
print("Transpose:\n ",transpose)
```
