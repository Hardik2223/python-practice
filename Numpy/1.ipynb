 ## CREATING ARRAYS FROM NUMPY#$
 

```python
arr_1d = np.array([1,2,3,4,5,6])
arr_2d = np.array([[1,2,3],[4,5,6]])
print("1D array: ",arr_1d)
print("2D array: ",arr_2d)
```

## list V/S  numpy arrays

```python
list1 = [1,2,3]
print("list multiplication:", list1 *2)

arry = np.array([1,2,3])
print("array multiplication:", arry *2)
```
